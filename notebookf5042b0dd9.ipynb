```python
import pandas as pd
import os

data_path = '/kaggle/input/olist-fact-for-tableau'

# смотрим файлы внутри
print(os.listdir(data_path))

# читаем все csv и склеиваем
dfs = []
for file in os.listdir(data_path):
    if file.endswith('.csv'):
        print("Читаю:", file)
        df_part = pd.read_csv(os.path.join(data_path, file))
        dfs.append(df_part)

df = pd.concat(dfs, ignore_index=True)

df.shape
```

```python
for file in os.listdir(data_path):
    df_temp = pd.read_csv(os.path.join(data_path, file))
    print(file, df_temp.shape)
```
